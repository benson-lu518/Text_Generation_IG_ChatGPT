## Set tokens from Instgram Graph API

```python
ACCESS_TOKEN = ""
INSTAGRAM_BUSINESS_USER_ID = ""
```

## Get the hashtag ID by hashtag

```python
import requests
from openai import OpenAI, OpenAIError

def get_id_by_hashtag(hashtag):
    """
    Fetch posts by a specific hashtag from Instagram API.
    """
    url = f"https://graph.facebook.com/v15.0/ig_hashtag_search"
    params = {
        'user_id': INSTAGRAM_BUSINESS_USER_ID,  # Instagram Business Account ID
        'q': hashtag,  # Hashtag to search for
        'access_token': ACCESS_TOKEN
    }
    
    response = requests.get(url, params=params)
    
    if response.status_code == 200:
        data = response.json()
        if 'data' in data and data['data']:
            hashtag_id = data['data'][0]['id']  # First match
            return hashtag_id
        else:
            print(f"No hashtag found for: {hashtag}")
            return None
    else:
        print(f"Error fetching hashtag ID: {response.status_code}")
        return None

hashtag_id = get_id_by_hashtag(hashtag)

```

```python
print(f"Hashtag ID for #{hashtag}: {hashtag_id}")
```

## Collect the top media posts by hashtag ID

```python
def get_posts_by_id(hashtag_id):
    url = f"https://graph.facebook.com/v15.0/{hashtag_id}/top_media"
    params = {
        'user_id': INSTAGRAM_BUSINESS_USER_ID,  
        'access_token': ACCESS_TOKEN,
        # 'fields': 'id,caption,comments_count,like_count,timestamp'
        'fields': 'caption'

    }
    
    response = requests.get(url, params=params)
    
    if response.status_code == 200:
        data = response.json()
        if 'data' in data and data['data']:
            posts = data['data']  # First match
            return posts
        else:
            print(f"No hashtag found for: {hashtag_id}")
            return None
    else:
        print(f"Error fetching hashtag ID: {response.status_code}")
        return None

posts = get_posts_by_id(hashtag_id)
caption = "\n".join(content['caption'] for content in posts) ##To str
```

## Generate article using gpt-3.5-turbo model and OpenAI API

```python
def generate_article(prompt,caption):
    # Create an OpenAI client instance
    client = OpenAI(api_key=OPEN_AI_KEY)

    # Create a chat completion
    try:
        completion = client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": "è«‹è¨˜ä½ï¼Œä»¥ä¸‹éƒ½è¦ç”¨ä¸­æ–‡å›ç­”."},
                {
                    "role": "user",
                    "content": prompt +"\n"+caption
                }
            ]
        )

        return completion.choices[0].message.content

    except OpenAIError as e:
        print(f"An error occurred: {e}")
article = generate_article(prompt,caption)
```

### Gunning Fog Index
The Gunning Fog Index measures the years of formal education needed to understand the text on a first reading. The scores are interpreted as follows:
- **6-8**: Easily readable for most people
- **9-12**: Fairly easy to read; suitable for high school students
- **13-16**: Requires some college education to understand
- **17+**: Very complex; likely understood by those with a graduate education

## 6. Topic Modeling

### Latent Dirichlet Allocation (LDA) model

Latent Dirichlet Allocation (LDA) is a generative statistical model used to classify text in a document to a particular topic. It assumes that each document is a mixture of topics and each topic is characterized by a distribution over words.

#### Steps to Implement LDA

1. **Preprocessing the Text**: 
   - Clean the text by removing stop words, punctuation, and any irrelevant information.
   - Tokenize the text into words.

2. **Creating the Dictionary and Corpus**:
   - Create a dictionary that maps every unique word to an integer ID.
   - Construct a corpus (a collection of documents) that reflects the frequency of words in each document.

3. **Building the LDA Model**:
   - Utilize the Gensim library to create and train the LDA model using the corpus and dictionary.

4. **Analyzing the Topics**:
   - Extract the main topics from the trained LDA model and identify the keywords associated with each topic.

### 1. **Preprocessing the Text**


```python
#Text mining package "ckip_transformers" from  Academia Sinica

from ckip_transformers.nlp import CkipWordSegmenter, CkipPosTagger, CkipNerChunker
import os
import re
import numpy as np
import pandas as pd
from collections import Counter

import warnings
warnings.filterwarnings("ignore")

# Initialize drivers
print("Initializing drivers ... WS")
ws_driver = CkipWordSegmenter(model="bert-base", device=-1) #-1: cpu
print("Initializing drivers ... POS")
pos_driver = CkipPosTagger(model="bert-base", device=-1)
print("Initializing drivers ... NER")
ner_driver = CkipNerChunker(model="bert-base", device=-1)
print("Initializing drivers ... all done")
print()
```

```python
# Part-of-speech tagging
def do_CKIP_POS(ws_result):
    pos = pos_driver(ws_result)
    all_list = []
    for sent in pos:

        all_list.append(sent)
    return all_list
```

```python
# Only remain Noun and Verb
def pos_filter(pos):
    for i in list(set(pos)):
        if i.startswith("N") or i.startswith("V"):
            return "Yes"
        else:
            continue
```

```python
# Remove number url
def remove_number_url(ws):
    number_pattern = "^\d+\.?\d*"
    url_pattern = "^https?:\\/\\/(?:www\\.)?[-a-zA-Z0-9@:%._\\+~#=]{1,256}\\.[a-zA-Z0-9()]{1,6}\\b(?:[-a-zA-Z0-9()@:%_\\+.~#?&\\/=]*)$"
    space_pattern = "^ .*"
    num_regex = re.match(number_pattern, ws)
    url_regex = re.match(url_pattern, ws)
    space_regex = re.match(space_pattern, ws)
    if num_regex ==  None and url_regex == None and space_regex == None:
        return True
    else:
        return False
```

```python
# Peprocessing
def cleaner(word_list, pos_results):
    word_lst = []
    for ws, pos in zip(word_list, pos_results):
        if_len_greater_than_1 = len(ws) > 1        #è©çµ„é•·åº¦å¿…é ˆå¤§æ–¼1
        is_V_or_N = pos_filter(pos[0])                #è©çµ„æ˜¯å¦ç‚ºåè©ã€å‹•è©
        is_num_or_url = remove_number_url(ws)      #è©çµ„æ˜¯å¦ç‚ºæ•¸å­—ã€ç¶²å€ã€ç©ºç™½é–‹é ­
        if if_len_greater_than_1 and is_V_or_N == "Yes" and is_num_or_url:
            word_lst.append(str(ws))
        else:
            pass
    return word_lst
```

```python
pos_results = do_CKIP_POS(word_list) 
word_final = [cleaner(word_list,pos_results)]
print(word_final)
```

### 2. **Create a dictionary and corput**


```python
import gensim
from gensim import corpora, models
from gensim.models.coherencemodel import CoherenceModel
from gensim.models.ldamodel import LdaModel
from pprint import pprint
```

```python
dictionary = corpora.Dictionary(word_final)
corpus = [dictionary.doc2bow(i) for i in word_final]
print(corpus) #(id, count)
```

### 3. **Build the LDA Model**


```python
def coherence(num_topics):
    ldamodel = LdaModel(corpus, num_topics = num_topics, id2word = dictionary, passes = 30, random_state = 42)
    # Top 5 words for each topic
    print(ldamodel.print_topics(num_topics = num_topics, num_words = 15))
    ldacm = CoherenceModel(model = ldamodel, texts = word_final, dictionary = dictionary, coherence="c_v")
    # Creates a coherence model to evaluate the quality of the topics generated by the LDA model.
    print(ldacm.get_coherence())
    return ldacm.get_coherence()
```

```python
def coherence_count(num_topics):
    x = range(1,num_topics+1)
    y = [coherence(i) for i in x]

    return x, y
    
x, y = coherence_count(20)
```

```python
# Get the number of topics with the highest coherence value
import matplotlib.pyplot as plt
import matplotlib

plt.plot(x,y)
plt.xlabel("Number of topics")
plt.ylabel("coherence")
plt.rcParams["font.sans-serif"] = ["Microsoft JhengHei"]
matplotlib.rcParams["axes.unicode_minus"] = False
plt.title("Topic-coherence changes")
plt.show()
```
